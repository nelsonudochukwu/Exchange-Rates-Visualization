```python
import pandas as pd

exchange_rates = pd.read_csv("C:\\Users\\NELSON ESTHER\\Desktop\\datasets for projects\\Exhange rates dataset\\euro-daily-hist_1999_2022.csv")
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar,', 'Period\\Unit:': 'Time'}, inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts() 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```
